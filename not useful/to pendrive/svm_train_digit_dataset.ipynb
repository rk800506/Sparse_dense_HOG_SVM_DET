```python
import pandas as pd 
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits['data']
```

```python
from matplotlib import pyplot as plt 
```

```python
plt.imshow(digits['images'][0], 'gray')
plt.show()
```

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df.head()
```

```python
df['target'] = digits.target
df.head()

df.head()
```

```python
df['digit_names'] = df.target.apply(lambda x: digits.target_names[x])
df.head()
```

```python
### extract data from df for train ing and testing ########

x = df.drop(['target','digit_names'], axis='columns')
x.head()
```

```python
### target classification classes ####
y = df.target
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, shuffle=True)
```

```python
len(x_train)
len(x_test)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(max_iter=10)
print(model)
```

```python
## train ##
model.fit(x_train, y_train)
```

```python
## test ###
model.score(x_test, y_test)
```
